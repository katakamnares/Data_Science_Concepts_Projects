### One Hot Encoder

#### Using sklearn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
data2=pd.read_csv("iris.csv")
data2.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data2
```

```python
# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore')
```

```python
enc.fit_transform(data2[['Species']]).toarray().astype('int')
```

```python
# passing bridge-types-cat column (label encoded values of bridge_types)
enc_df = pd.DataFrame(enc.fit_transform(data2[['Species']]).toarray().astype('int'),columns=['Setosa','Versicular','Verginica'])
```

```python
enc_df
```

```python
# merge with main df
data_final = data2.iloc[:,0:4].join(enc_df)
data_final
```

#### IsolationForest

```python
from sklearn.ensemble import IsolationForest
```

```python
data =pd.read_csv("iris.csv",index_col=0)
data_encoded=pd.get_dummies(data,dtype=int)
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(data_encoded)
```

```python
# predictions
y_pred_outliers = clf.predict(data_encoded)
```

```python
#-1 for outliers and 1 for inliers.
y_pred_outliers
```

```python
data_encoded
```

```python
## Let us add a new data point which is outlier
data_encoded.loc[151]=[20,40,30,50,1,0,0]
```

```python
data_encoded
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(data_encoded)
# predictions
y_pred_outliers = clf.predict(data_encoded)
y_pred_outliers
```

```python
data_encoded['scores']=clf.decision_function(data_encoded)
```

```python
data_encoded['anomaly']=clf.predict(data_encoded.iloc[:,0:7])
```

```python
data_encoded
```

```python
#Print the outlier data points
data_encoded[data_encoded['anomaly']==-1].index.values
```
