```python
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
df=sn.load_dataset('tips')
sn.stripplot()
sn.swarmplot(x='day',y='tip',data=df,hue='sex')
plt.show()
```

```python
plt.subplot(1,2,1)
sn.violinplot(x='day',y='total_bill',data=df)
plt.ylabel('tip ($)')
plt.subplot(1,2,2)
sn.boxplot(x='time',y='tip',data=df)
plt.ylabel("bill")
plt.show()
```
