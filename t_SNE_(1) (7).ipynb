```python
from pandas import read_csv
import pandas as pd

from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
```

```python
# load data
filename = 'TSNE_data.csv'
dataframe = read_csv(filename)
```

```python
dataframe.info()
```

```python
dataframe.head()
```

```python
# Split-out validation dataset
array = dataframe.values
# separate array into input and output components
X = array[:,1:]
Y = array[:,0]
```

```python
X
```

```python
#TSNE visualization
from bioinfokit.visuz import cluster

data_tsne = TSNE(n_components=2,perplexity=50).fit_transform(X)
data_tsne
```

```python
cluster.tsneplot(score=data_tsne)
```

```python
# get a list of categories
color_class = dataframe['diagnosis'].to_numpy()
print(color_class)
#Plot will be stored in the default directory
```

```python
cluster.tsneplot(score=data_tsne, colorlist=color_class, legendpos='upper right',legendanchor=(1.15, 1))
```

```python
data=pd.DataFrame(data=data_tsne,columns=['var1','var2'])
data.head()
```

```python
data['diagonosis']=dataframe['diagnosis']
```

```python
data.head()
```

```python
df=pd.get_dummies(data,dtype=int,drop_first=True)
```

```python
df
```

```python
X=df.drop('diagonosis_M',axis=1)
y=df['diagonosis_M']
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X,y)
```

```python
pred=lr.predict(X)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,pred)
```
